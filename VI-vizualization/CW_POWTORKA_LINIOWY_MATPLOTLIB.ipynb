```python
df = pd.read_csv('data/temperature.csv')
df
```

```python
## przefiltruj dane dla Xian oraz dla okresu mniejszego niz 1822-01-01
## zrob to dla zmiennej df_xian (może być w jednej linijce albo w dwóch :)
## za pomoca loca albo poprzez zwykle przypisanie series to series dokonaj zastapienia wartosci AverageTemperature i AverageTemperatureUncertainty
## wartościami skróconymi do jednego miejsca po kropce :) poprzez
## series.round c
```

```python
df_xian = df.loc[ (df['City'] == 'Xian') & (df['dt'] < '1822-01-01')]
```

```python
df_xian = df_xian.rename(columns={'AverageTemperature':'3'})
df_xian = df_xian.rename(columns={'AverageTemperatureUncertainty':'AverageTemperature'})
df_xian = df_xian.rename(columns={'3':'AverageTemperatureUncertainty'})
```

```python
df_xian
```

```python
df_xian['AverageTemperature'] = df_xian['AverageTemperature'].round(decimals=1)
df_xian['AverageTemperatureUncertainty']= df_xian['AverageTemperatureUncertainty'].round(decimals=1)
df_xian
```
