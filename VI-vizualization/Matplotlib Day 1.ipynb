```python
#How to plot 2 series on one plot. 
import matplotlib.dates as mdates

afig, ax = plt.subplots(figsize=(10,5))


ax.set_title('Sell Corelation')
ax.set_xlabel('Time', rotation=45)
ax.set_ylabel('Sell', rotation=45)

xformatter = mdates.DateFormatter('%H:%M')
plt.gcf().axes[0].xaxis.set_major_formatter(xformatter)

plt.xticks(rotation=45, ha='right')

# series dla sprzedaz 1
ax.plot(dane['DT'], dane['Sprzedaz'])
ax.plot(dane['DT'], dane['Sprzedaz2'], linewidth=2, linestyle='--', color='r')


for i, j in zip(dane['DT'], dane['Sprzedaz']):
    ax.annotate(str(j), xy=(i,j))

# series dla sprzedaz 2
for i, j in zip(dane['DT'], dane['Sprzedaz2']):
    ax.annotate(str(j), xy=(i,j))
    

# stworzeneie legendy 
ax.legend(['Sprzedaz', 'Sprzedaz2'])
```

```python
df = pd.DataFrame({
    'Male': [35, 20, 43, 12, 19],
    'Female': [20, 11, 18, 6, 9],
    'Not sure': [20, 11, 18, 6, 9]
 })

Class = ["Fisrt","Second","Third","Fourth","Fifth"]

ax = df.plot(stacked=True, kind='bar')

for bar in ax.patches:
    print(bar)
    height = bar.get_height()
    width = bar.get_width()
    
    x = bar.get_x()
    y = bar.get_y()
    
    label_text = height
    
    label_x = x + width/2
    label_y = y + height/2
    
    ax.text(label_x, label_y, label_text, ha='center', va='center', color='white')
    
ax.set_xticklabels(Class, rotation = 'horizontal')
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.3, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()

ax.pie(
# specific vlaues of chart
    sizes,
       
# to add labels
      labels=labels,

# argument to make it further.
      explode=explode,
      
# to add values
      autopct='%1.0f%%',
    
# size of chart
    radius=0.99
      )
```

```python
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = 'lines',
                    name = 'Cytowania',
                    marker = dict(color='rgba(16,112,2,0.8)'),
                    text = df.university_name
)

trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = 'lines+markers',
                    name = 'Poziom nauki',
                    marker = dict(color='rgba(80,26,80,0.8)'),
                    text = df.university_name
)
data = [trace1,trace2]
layout = dict(title="Poziom nauki i cytowania, a pozycja w rankingu",
             xaxis=dict(title="Ranking wiatowy",
                       zeroline = True))
fig = dict(data=data, layout=layout)
iplot(fig)
```
