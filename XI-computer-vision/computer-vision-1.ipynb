```python
import cv2
import matplotlib.pyplot as plt
```

```python
spongeboob = cv2.imread('data/SpongeBob_SquarePants_character.jpg', cv2.IMREAD_UNCHANGED)
```

```python
def cv2_imshow(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
cv2_imshow(spongeboob)
```
