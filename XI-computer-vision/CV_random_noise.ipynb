```python
import cv2
# from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
# wczytanie obrazu
img = cv2.imread('ML-datasets/images/gory_lc.png')
# od razu - zamiana na RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img = img / 255
```

```python
# szum gaussowski - Å›rednia 0, std 1/10
random_noise = np.random.randn(*img.shape) / 10
```

```python
np.mean(random_noise), np.std(random_noise)
```

```python
# dodanie szumu do obrazu
noisy_img = np.add(img, random_noise)
```

```python
# obraz z szumem
plt.imshow(noisy_img) 
```

```python
# oryginalny obraz
plt.imshow(img)
```
