**Remember Linear and Logistic regression?**



Did you know that those two models can be expressed as a single neuron neural network?




To create a **Linear Regression** model create a model with one `Dense` layer and set the `units` parametr to `1`. That's it!






To create a **Logistic Regression** model create the same model but set another parameter - `activation = "sigmoid"`.



# Linear Regression

```python
import tensorflow as tf
import pandas as pd

# create fake dataset for regression using random tensors we used in the first exercise
# make it a 100 samples for training, 30 for validation and 30 for testing,
# a sample should contain 5 features - X_train (100, 5), X_val, X_test (30,5)

X_train = tf.random.normal((100, 5))
X_val = tf.random.normal((30, 5))
X_test = tf.random.normal((30, 5))

y_train = tf.random.normal((100, 1))
y_val = tf.random.normal((30, 1))
y_test = tf.random.normal((30, 1))

# define your LinearRegression model as a Sequential model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=(5,1)))

model.compile(loss='mse', metrics=['mae'], optimizer='sgd')

# train it ... 
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val))
# ... and evaluate,
model.evaluate(X_test, y_test)

# now plot the training curves
df = pd.DataFrame(history.history).plot()
```

# Logistic regression

```python
import tensorflow as tf
import pandas as pd

# create fake dataset for regression using random tensors we used in the first exercise
# make it a 100 samples for training, 30 for validation and 30 for testing,
# a sample should contain 5 features - X_train (100, 5), X_val, X_test (30,5)

X_train = tf.random.normal((100, 5))
X_val = tf.random.normal((30, 5))
X_test = tf.random.normal((30, 5))

y_train = tf.random.normal((100, 1))
y_val = tf.random.normal((30, 1))
y_test = tf.random.normal((30, 1))

# define your LinearRegression model as a Sequential model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=(5,1), activation='sigmoid'))

model.compile(loss='mse', metrics=['accuracy'], optimizer='adam')

# train it ... 
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val))
# ... and evaluate,
model.evaluate(X_test, y_test)

# now plot the training curves
df = pd.DataFrame(history.history).plot()
```
